```python
import sys, platform, os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

camb_path = os.path.realpath(os.path.join(os.getcwd(),'..'))
sys.path.insert(0,camb_path)
import camb
from camb import model, initialpower
```

Read the Planck CMB power spectrum data from web.

```python
url = 'https://github.com/mengkel/CMB-power-spectrum/blob/main/power-data.csv'
data = pd.read_html(url)[0]
```

Set up a new set of parameters for CAMB. One can modify any parameters to see the difference.

```python
pars = camb.CAMBparams()
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122, mnu=0.06, omk=0, tau=0.06)
pars.InitPower.set_params(As=2e-9, ns=0.965, r=0)
pars.set_for_lmax(2500, lens_potential_accuracy=0)
pars.set_dark_energy(w=-1)
results = camb.get_results(pars)
```

Calculate results for these parameters and select the data you need.

```python
powers =results.get_cmb_power_spectra(pars, CMB_unit='muK')
totCL=powers['total']
ls = np.arange(totCL.shape[0])
```

Now, plot the CMB power spectrum data and compare with the model.  

```python
plt.plot(ls,totCL[:,0], label = 'fit line')
plt.errorbar(data['l'],data['Dl'], label = 'spectrum data and error bar', yerr=[data['dlm'],data['dlp']], xerr=None,  fmt = 'o', markersize=3, ecolor='r',color='b',elinewidth=1,capsize=2)
plt.xlim(1.9, 2500)
plt.ylim(0, 6000)
plt.xscale('log')
plt.ylabel(r'$l(l+1)C_l/2\pi$ [$\mu K^2$]')
plt.xlabel('Multipole l')
plt.title('CMB spectrum: power spectrum')
plt.legend()
```
